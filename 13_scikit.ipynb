### Gini Impurity

* Memiliki jangkauan nilai antara 0 dan 1
* 0 = nilai murni sempurna
* 1 = nilai paling tidak murni

**Rumus:**
G = 1 - (total hasil penjumlahan probability kuadrat dari tiap kelas yang tersedia)

### Information Gain

Information Gain adalah selisih antara nilai gini impurity dikurang average gini impurity yang telah di splitting

### Implementasi dengan Scikit Learn

```python
from sklearn.datasets import load_iris
X, y = load_iris(return_X_y=True)

print(f'Dimensi Feature: {X.shape}')
print(f'Class: {set(y)}')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.3,
                                                    random_state=0)
```

```python
#Classification dengan DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=4)
model.fit(X_train, y_train)
```

```python
#Visualisasi Model
import matplotlib.pyplot as plt
from sklearn import tree

plt.rcParams['figure.dpi'] = 85
plt.subplots(figsize=(10, 10))
tree.plot_tree(model, fontsize=10)
plt.show()
```

```python
#Evaluasi Model
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```
